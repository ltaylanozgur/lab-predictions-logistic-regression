## Read the data into a Pandas dataframe.

```python
# get the data
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

query = '''SELECT * FROM sakila.film f
LEFT JOIN sakila.inventory i USING (film_id)
LEFT JOIN sakila.rental fa USING (inventory_id)
LEFT JOIN sakila.film_category fc USING (film_id)
LEFT JOIN sakila.category c USING (category_id)
LEFT JOIN sakila.payment p USING (rental_id);'''

sakila = pd.read_sql_query(query, engine)
sakila.head()
```

```python
sakila.shape
```

```python
sakila.dtypes
```

```python
#Drop columns
columns_raw=list(sakila.columns)

sakila.dropna(axis=1, thresh= 0.80*len(sakila), inplace=True) 
#thresh = x * len(DataFrame) => x: ratio of clean data to all data in rows of the dataframe. 
#The remaining is NaN value. 

columns_result=list(sakila.columns)

columns_dropped=list(set(columns_raw) - set(columns_result))

print(columns_dropped)
```

```python
sakila['rental_date'] = pd.to_datetime(sakila['rental_date'], errors='coerce')
```

```python
sakila['rental_status'] = pd.DatetimeIndex(sakila['rental_date']).month
```

```python
#Films rented in May is represented by 1.
#Films not rented in May is represented by 0.

def ren_stat(x):
    if x==5:
        return 1
    else:
        return 0
```

```python
sakila['rental_status'] = list(map(ren_stat, sakila['rental_status']))
```

```python
sakila.head(60)
```

```python
sakila.dtypes
```

```python
# set duration_of_rental column

sakila['duration_of_rental'] = (sakila['return_date'] - sakila['rental_date'])

sakila.head()

sum(sakila['duration_of_rental'].isna())

# Replace the NaNs with the column median 

time_median = sakila['duration_of_rental'].median()
time_median
sakila['duration_of_rental'] = sakila['duration_of_rental'].replace({np.nan: time_median}).dt.days
```

```python
#Drop columns
def columns_nan(data, list):
    
    for a in list:
        data.drop(a, axis=1, inplace=True) #drop columns
    
    return data
```

```python
#Drop columns
cols_drop = ['staff_id','store_id','rental_duration','category_id', 'rental_id', 'inventory_id', 'release_year', 'language_id',
                     'last_update', 'rental_date', 'customer_id', 'return_date', 'replacement_cost', 'payment_id', 'payment_date', 'special_features']     
```

```python
#Drop columns
sakila = columns_nan(sakila,cols_drop)
```

```python
sakila['title'].value_counts()
```

```python
#Duplicates are dropped based on title.
sakila.drop_duplicates(subset=['film_id', 'rental_status'],inplace=True)
```

```python
sakila.sort_values(by=['rental_status'],ascending=False,inplace=True)
sakila.head(10)
sakila.drop_duplicates(subset='title', keep='first', inplace=True) #duplicates are gone!

sakila.shape

sakila['rental_status'].value_counts(dropna=False)
```

```python
# look for multicolinearity (some columns having almost identical correlation to other variables)
corr_matrix=sakila.corr(method='pearson')  # default
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

### X-y Split

```python
# Need to X-y-split AND train-test-split BEFORE I apply transformations, 
# then train transformation on training set only
y = sakila['rental_status']
X = sakila.drop('rental_status', axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

#### Normalizing data: make data range from 0 - 1, instead of from min to max

```python
X_train_num = X_train.select_dtypes(include = np.number)
X_train_num.head()
```

```python
transformer = MinMaxScaler().fit(X_train_num)
X_train_normalized = transformer.transform(X_train_num)
X_train_normalized_scale = pd.DataFrame(X_train_normalized,columns=X_train_num.columns)
X_train_normalized_scale.head()
```

#### Encoding categoricals

```python
X_train_cat = X_train.select_dtypes(include = object)
X_train_cat.head()
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_cat)
cols = encoder.get_feature_names(input_features=X_train_cat.columns)
```

```python
print(encoder.categories_)
```

```python
encoded = encoder.transform(X_train_cat).toarray()
print(encoded)
```

```python
onehot_encoded = pd.DataFrame(encoded,columns=cols)
onehot_encoded.head(5)
```

### CONCAT Minmax and OneHotEncoded

```python
print(X_train_normalized_scale.shape,onehot_encoded.shape)
```

```python
# let's merge all this information together into a single dataset with all features, now numerical
X_train_transformed = pd.concat([X_train_normalized_scale, onehot_encoded], axis=1)
```

```python
print(X_train_transformed.shape,y_train.shape)
```

## Create a logistic regression model to predict this variable from the cleaned data.

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='multinomial').fit(X_train_transformed, y_train)

#classifications allow me to predict.
```

```python
# for numericals
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
X_test_norm.shape
```

```python
# for categoricals
X_test_cat = X_test.select_dtypes(include = np.object)

#Encoding categoricals using previous encoder
#We do not need to fit again.
encoded = encoder.transform(X_test_cat).toarray()
cols = encoder.get_feature_names(input_features=X_test_cat.columns)
onehot_encoded = pd.DataFrame(encoded,columns=cols)
```

```python
print(encoder.categories_)
```

```python
print(encoded)
```

```python
onehot_encoded.head(5)
```

```python
print(X_test_norm.shape,onehot_encoded.shape)
```

```python
X_test_transformed = np.concatenate([X_test_norm, onehot_encoded], axis=1)
```

```python
predictions = classification.predict(X_test_transformed)
predictions
classification.score(X_test_transformed, y_test)
```

```python
print(y_test.value_counts())
```

```python
pd.Series(predictions).value_counts()
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)

# number of times of predictions

# predicted   |  0 | 1  |
# --------------------------
# actual    0 |  0 | 61 |
# --------------------------
#           1 |  0 | 139|   
# --------------------------
```

```python
# KNN classifier: look at nearest neighbours and use the majority to determine class
from sklearn import neighbors
clf = neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform')
clf.fit(X_train_transformed, y_train)
predictions_clf = clf.predict(X_test_transformed)
clf.score(X_test_transformed, y_test)
```

```python
confusion_matrix(y_test, predictions_clf)
```

```python
# predicted   |    0  |  1 |
# --------------------------
# actual    0 |   13  | 48 |
# --------------------------
#           1 |   28  | 111|   
# --------------------------
```

```python
pd.Series(y_test).value_counts()
```

```python
pd.Series(predictions_clf).value_counts()
```

## Evaluate the results. 

The actual data shows that 686 films are rented in May and 314 films are not rented in May out of 1000 different films. Our model predicts that 200 films out of 200 tested films are going to be rented in May. However, 139 of them are rented in May. The logistic regression model gives 0.695 score. Although KNN classifier has lower score (0.625), confusion matrix show that 28 films are rented out of 159 films that are predicted by the KNN classifier to be rented in May. As a result, KNN classifier relatively predicts better than the logistic regression model.

The May data is subdivided into two: 314 films that are rented in May and 686 films that are not rented in May. This data implies an imbalanced data. Our model could be improved if either oversampling, undersamling or SMOTE is applied.
